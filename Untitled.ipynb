# Skill Normalization Functions

```python
def normalize_skills(skills):
    skills = skills.lower()
    skills = re.split(r'[,\s]+', skills)
    normalized_skills = []
    for skill in skills:
        if skill in SKILL_ALIASES:
            normalized_skills.append(SKILL_ALIASES[skill])
        elif ' ' in skill:
            for alias, value in SKILL_ALIASES.items():
                if alias == skill:
                    normalized_skills.append(value)
                    break
    return list(set(normalized_skills))

for resume_id, resume in resume_dataset.items():
    skills = resume["skills"]
    normalized_skills = normalize_skills(skills)
    resume_dataset[resume_id]["normalized_skills"] = normalized_skills
```

```python
def computetfidf(resume_dataset):
    vocabulary = set()
    for resume in resume_dataset.values():
        vocabulary.update(resume["normalized_skills"])
    vocabulary = sorted(list(vocabulary))

    tfidfvectors = {}
    for resumeid, resume in resume_dataset.items():
        tfidfvector = [0] * len(vocabulary)
        for i, skill in enumerate(vocabulary):
            if skill in resume["normalized_skills"]:
                tf = 1 / len(resume["normalized_skills"])
                df = sum(1 for r in resume_dataset.values() if skill in r["normalizedskills"])
                idf = math.log(10 / df)
                tf_idf = tf * idf
                tfidfvector[i] = tf_idf
        tfidfvectors[resumeid] = tfidfvector
    return tfidfvectors, vocabulary

tfidfvectors, vocabulary = computetfidf(resume_dataset)
```
